```python
import pandas as pd
```

```python
# Examen final

# Ejercicio propuesto integrador: Limpieza de datos

csvAbs = pd.read_csv("Absenteeism-data.csv")
dfAbs = csvAbs.copy()
dfAbs
```

```python
# 1. Elimina la columna "ID":

dfAbs = dfAbs.drop(columns=['ID'], errors='ignore')
dfAbs
```

```python
# 2. Divide las "Razones de ausencia" en mÃºltiples variables
#    dummy y luego agrÃºpalas de la siguiente manera:

dummiesRazAus = pd.get_dummies(dfAbs['Reason for Absence'], dtype='uint8')
dummiesRazAus
```

```python
# # Grupo 1: Columnas 1 a 14

colGrupo1 = dummiesRazAus.loc[:, 1:14].max(axis=1)
colGrupo1
```

```python
# # Grupo 2: Columnas 15, 16 y 17

colGrupo2 = dummiesRazAus.loc[:, 15:17].max(axis=1)
colGrupo2
```

```python
# # Grupo 3: Columnas 18, 19, 20 y 21

colGrupo3 = dummiesRazAus.loc[:, 18:21].max(axis=1)
colGrupo3
```

```python
# # Grupo 4: Columnas 22 a 28

colGrupo4 = dummiesRazAus.loc[:, 22:28].max(axis=1)
colGrupo4
```

```python
# # Agrupar:

dfAbs = pd.concat([colGrupo1, colGrupo2, colGrupo3, colGrupo4, dfAbs], axis=1)
nomColumnas = [
    'Razones 1', 'Razones 2', 'Razones 3', 'Razones 4', 
    'Reason for Absence', 'Date', 'Transportation Expense', 
    'Distance to Work', 'Age', 'Daily Work Load Average', 
    'Body Mass Index', 'Education', 'Children', 'Pets', 
    'Absenteeism Time in Hours']
dfAbs.columns = nomColumnas
dfAbs
```

```python
# 3. Eliminar la columna "Reason for Absence":

dfAbs = dfAbs.drop(columns=['Reason for Absence'], errors='ignore')
dfAbs
```

```python
# 4. Extrae el valor del mes y del dÃ­a de la semana a partir
#    de la columna "Date". Luego, elimina tambiÃ©n la columna "Date":

dfAbs['Date'] = pd.to_datetime(dfAbs["Date"], format='%d/%m/%Y')

# # Columna "Mes":

listMeses = []

for i in range(dfAbs.shape[0]):
    listMeses.append(dfAbs['Date'][i].month)

dfAbs['Mes'] = listMeses
dfAbs
```

```python
# # Columna "Dia de semana":

dfAbs['Dia de semana'] = dfAbs['Date'].dt.weekday
dfAbs
```

```python
# # Eliminar la columna "Date":

dfAbs = dfAbs.drop(columns=['Date'], errors='ignore')
dfAbs
```

```python
# # Reordenamos la tabla:

nomColumnas = [
    'Razones 1', 'Razones 2', 'Razones 3', 'Razones 4', 
    'Mes', 'Dia de semana', 'Transportation Expense', 
    'Distance to Work', 'Age', 'Daily Work Load Average', 
    'Body Mass Index', 'Education', 'Children', 'Pets', 
    'Absenteeism Time in Hours']
dfAbs = dfAbs[nomColumnas]
dfAbs
```

```python
# 5. Convierte los datos de la columna "Education" en datos binarios, 
#    mapeando el valor "0" a "1", y el valor "1" y el resto de valores 
#    encontrados en esta columna a "0":

dfAbs['Education'].value_counts()
```

```python
dfAbs['Education'] = dfAbs['Education'].map({1:0, 2:1, 3:1, 4:1})
dfAbs
```
