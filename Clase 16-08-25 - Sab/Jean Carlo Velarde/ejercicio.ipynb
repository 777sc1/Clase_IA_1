```python
import pandas as pd
```

```python
fileCsv = "Absenteeism-data.csv"
dataFrame = pd.read_csv(fileCsv)
dataFrame.head()
```

```python
dataFrame = dataFrame.drop(columns=["ID"])
dataFrame.head()
```

```python
dummy = pd.get_dummies(dataFrame['Reason for Absence'], prefix='Grupo', drop_first=False)
dataFrame['Grupo_1'] = dummy .loc[:, 'Grupo_1':'Grupo_14'].max(axis=1).astype(int)
dataFrame['Grupo_2'] = dummy .loc[:, 'Grupo_15':'Grupo_17'].max(axis=1).astype(int)
dataFrame['Grupo_3'] = dummy .loc[:, 'Grupo_18':'Grupo_21'].max(axis=1).astype(int)
dataFrame['Grupo_4'] = dummy .loc[:, 'Grupo_22':'Grupo_28'].max(axis=1).astype(int)

dataFrame = dataFrame.drop(columns=['Reason for Absence'])

dataFrame.head()
```

```python
dataFrame['Date'] = pd.to_datetime(dataFrame['Date'], dayfirst=True)
dataFrame['Month'] = dataFrame['Date'].dt.month
dataFrame['Day of the Week'] = dataFrame['Date'].dt.weekday
dataFrame = dataFrame.drop(columns=['Date'])
dataFrame.head()
```

```python
dataFrame['Education'] = dataFrame['Education'].apply(lambda x: 1 if x == 0 else 0)
dataFrame.head()
```

```python
output = "df-cleaned.csv"
dataFrame.to_csv(output, index=False)
```
