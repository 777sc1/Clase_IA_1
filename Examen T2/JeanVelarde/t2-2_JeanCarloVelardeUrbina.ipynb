```python
import pandas as pd
import numpy as np
df_original = pd.read_csv('bios.csv')
df_original.head()
```

```python
for col in ['Used name', 'Full name']:
    if col in df_original.columns:
        df_original[col] = df_original[col].str.replace('•', '', regex=False).str.strip()

df_original.head()
```

```python
df_original[["Height", "Weight"]] = df_original["Measurements"].str.extract(r"([^/]+)\s*/\s*([^/]+)")
df_original["Height"] = df_original["Height"].str.strip()
df_original["Weight"] = df_original["Weight"].str.strip()
df_original[["Measurements", "Height", "Weight"]].head()

df_original.head()
```

```python
for col in ['Born', 'Died']:
    if col in df_original.columns:
            df_original[col + '_date'] = pd.to_datetime(
            df_original[col].str.extract(r'(\d{1,2}\s+\w+\s+\d{4})')[0],
            errors='coerce',
            format='%d %B %Y'
        )
df_original.head()
```

```python
if 'Born' in df_original.columns:
    place_parts = df_original['Born'].str.extract(r'in\s+(.+)$')[0].str.split(',', expand=True)
if place_parts.shape[1] >= 3:
    df_original['Born_city'] = place_parts[0].str.strip()
    df_original['Born_region'] = place_parts[1].str.strip()
    df_original['Born_country'] = place_parts[2].str.strip()
elif place_parts.shape[1] == 2:
    df_original['Born_city'] = place_parts[0].str.strip()
    df_original['Born_country'] = place_parts[1].str.strip()
df_original.head()
```

```python
cols_drop_bios = [
    'Measurements', 'Affiliations', 'Nick/petnames', 'Title(s)',
    'Other names', 'Original name', 'Name order'
]
df_original.drop(columns=[c for c in cols_drop_bios if c in df_original.columns], inplace=True)
df_original.head()
```

```python
df_original['Height_cm'] = df_original['Height_cm'].astype(float)
df_original['Weight_kg'] = df_original['Weight_kg'].astype(float)
df_original.to_csv('bios_JeanCarloVelardeUrbina.csv', index=False, encoding='utf-8-sig')
df_original.head()
```

```python
df_results = pd.read_csv('results.csv')
df_results.head()
```

```python
if 'Games' in df_results.columns:
    df_results['Year'] = df_results['Games'].str.extract(r'(\d{4})').astype(float).astype('Int64')
    df_results['Season'] = df_results['Games'].str.extract(r'\d{4}\s+(\w+)')
df_results.head()
```

```python
if 'Pos' in df_results.columns:
    df_results['Tied'] = df_results['Pos'].astype(str).str.contains('=', regex=False)
    df_results['Pos'] = pd.to_numeric(df_results['Pos'].astype(str).str.replace('=', '', regex=False), errors='coerce')
df_results.head()
```

```python
df_results = df_results.loc[:, ~df_results.columns.str.contains('^Unnamed')]
df_results.head()
```

```python
column_order = [
    'Year', 'Season', 'Games', 'Event', 'Team', 'Pos', 'Tied', 'Medal',
    'As', 'athlete_id', 'NOC', 'Discipline', 'Nationality'
]
df_results = df_results[[c for c in column_order if c in df_results.columns]]
df_results.head()
```

```python
df_results.to_csv("results_JeanCarloVelardeUrbina.csv", index=False, encoding="utf-8-sig")

df_results.head()
```
