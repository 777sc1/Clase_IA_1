```python
import pandas as pd
import numpy as np
```

```python
# Cargar el dataset
ruta = 'Top_1000_IMDb_movies_New_version.csv'
df = pd.read_csv(ruta)
```

```python
# Mostrar las primeras 5 filas
display(df.head())
```

```python
# Mostrar resumen de informaci√≥n
df.info()
```

```python
# Mostrar resumen estad√≠stico
display(df.describe(include="all"))
```

```python
# Renombrar columnas
df.columns = ["index", "title", "year", "duration", "rating", "metascore", "gross", "votes", "description"]
# Identificar valores nulos por columna
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
# Conversi√≥n de tipos de datos
df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")
df["votes"] = df["votes"].str.replace(",", "").astype("int")
```

```python
print("\nTipos de datos despu√©s de la conversi√≥n:")
print(df[["year", "duration", "rating", "votes"]].dtypes)
print("\nValores nulos en columnas num√©ricas clave:")
print(df[["year", "duration", "rating", "votes"]].isnull().sum())
```

```python
# Eliminar registros sin el a√±o
df = df.dropna(subset=["year"])
```

```python
# Rating
rating_mean = np.mean(df["rating"])
rating_median = np.median(df["rating"])
rating_min = np.min(df["rating"])
rating_max = np.max(df["rating"])
rating_range = np.ptp(df["rating"])

# Votes
votes_mean = np.mean(df["votes"])
votes_median = np.median(df["votes"])
votes_min = np.min(df["votes"])
votes_max = np.max(df["votes"])
votes_range = np.ptp(df["votes"])
```

```python
# Mostrar resultados
print("=== Movie Rating ===")
print(f"Media: {rating_mean}")
print(f"Mediana: {rating_median}")
print(f"M√≠nimo: {rating_min}")
print(f"M√°ximo: {rating_max}")
print(f"Amplitud: {rating_range}")

print("\n=== Votes ===")
print(f"Media: {votes_mean}")
print(f"Mediana: {votes_median}")
print(f"M√≠nimo: {votes_min}")
print(f"M√°ximo: {votes_max}")
print(f"Amplitud: {votes_range}")
```

```python
df["decade"] = (df["year"] // 10) * 10
```

```python
movies_per_decade = df.groupby("decade")["title"].count()
print("\nCantidad de pel√≠culas por d√©cada:")
print(movies_per_decade)
```

```python
rating_per_decade = df.groupby("decade")["rating"].mean().round(2)
print("\nCalificaci√≥n promedio por d√©cada:")
print(rating_per_decade)
```

```python
top10_movies = df.sort_values(by="rating", ascending=False)[["title", "year", "rating"]].head(10)
print("\nTop 10 pel√≠culas mejor valoradas:")
print(top10_movies.to_string(index=False))
```
