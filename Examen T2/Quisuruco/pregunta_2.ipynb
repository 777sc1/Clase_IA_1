```python
import pandas as pd
import numpy as np
```

```python
bios = pd.read_csv("bios.csv")
```

```python
display(bios.head())
```

```python
#Quitar • en Used name
bios["Used name"] = bios["Used name"].str.replace("•", "", regex=False).str.strip()
display(bios[["Used name"]])
```

```python
# Extraer altura y peso
bios[["Height", "Weight"]] = bios["Measurements"].str.extract(r"([^/]+)\s*/\s*([^/]+)")
bios["Height"] = bios["Height"].str.strip()
bios["Weight"] = bios["Weight"].str.strip()
display(bios[["Measurements", "Height", "Weight"]].head())
```

```python
# Extraer la fecha de Born y Died
bios["Born Date"] = bios["Born"].str.extract(r"^(.+?)\s+in\s")
bios["Died Date"] = bios["Died"].str.extract(r"^(.+?)\s+in\s")
display(bios[["Born","Died","Born Date", "Died Date"]].head())
```

```python
# Extraer ciudad, región y país de Born
bios[["Born Date", "Born City", "Born Region", "Born Country"]] = bios["Born"].str.extract(r"^([\dA-Za-z\s]+)\s+in\s+([^,]+),\s*([^(]+)\s*\(([^)]+)\)")
display(bios[["Born City", "Born Region", "Born Country"]].head())
```

```python
#Columna que va quedar para el guardado
cols_to_keep = [
    "Used name", "Height", "Weight",
    "Born Date", "Died Date",
    "Born City", "Born Region", "Born Country"
]
bios = bios[cols_to_keep]

# Guardar el CSV
bios.to_csv("bios.csv", index=False, encoding="utf-8-sig")

# Mostrar para verificar
display(bios.head())
```

```python
results = pd.read_csv("results.csv")
```

```python
display(results.head(10))
```

```python
#separamos el año y el tipo de games
results[["Games Year", "Games Type"]] = results["Games"].str.extract(r"(\d{4})\s*(.*)")
display(results[["Games", "Games Year", "Games Type"]].head())
```

```python
#Crear columna 'Tied' y extraemos los empates'
results["Tied"] = results["Pos"].astype(str).str.startswith("=").astype(int)
results["Pos"] = results["Pos"].astype(str).str.replace("=", "", regex=False)
#Convertir en NaN los valores no numéricos en la columna Pos
results["Pos"] = pd.to_numeric(results["Pos"], errors="coerce") 
display(results[["Pos", "Tied"]].head())
```

```python
# Eliminar columnas Unnamed
results = results.loc[:, ~results.columns.str.contains("Unnamed: 7")]
```

```python
column_order = ["Games Year", "Games Type", "Event", "Team", "Pos", "Tied", "Medal", 
                "As", "athlete_id", "NOC", "Discipline", "Nationality"]
results = results[column_order]
#Guardar CSV limpio
results.to_csv("results.csv", index=False, encoding="utf-8-sig")
# Verificar  filas
display(results.head())
```
