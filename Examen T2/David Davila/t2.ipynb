```python
import pandas as pd
import numpy as np
```

```python
# 1.
df = pd.read_csv('Top_1000_IMDb_movies_New_version.csv')

# Primeras 5 filas
print("Primeras 5 filas:")
display(df.head())

# Resumen de info
print("\nResumen info():")
df.info()

# Resumen estadístico
print("\nResumen describe():")
display(df.describe(include='all'))
```

```python
# 2.
# Revisar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())

# Renombrar columnas para evitar espacios y uniformidad
df.columns = df.columns.str.strip().str.replace(" ", "_").str.lower()

# Conversión de tipos
df['year_of_release'] = pd.to_numeric(df['year_of_release'], errors='coerce', downcast='integer')
df['watch_time'] = pd.to_numeric(df['watch_time'], errors='coerce', downcast='integer')
df['movie_rating'] = pd.to_numeric(df['movie_rating'], errors='coerce')
df['votes'] = df['votes'].replace({',':''}, regex=True)  # quitar comas en votos
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')

# Revisión final de tipos
print("\nTipos después de conversión:")
print(df.dtypes)

# Imputar nulos en columnas numéricas con la media
for col in ['watch_time', 'movie_rating', 'votes']:
    if df[col].isnull().any():
        df[col].fillna(df[col].mean(), inplace=True)


# 3.
ratings = df['movie_rating'].to_numpy()
votes = df['votes'].to_numpy()

print("\n=== Rating ===")
print(f"Media: {np.mean(ratings):.2f}")
print(f"Mediana: {np.median(ratings):.2f}")
print(f"Mínimo: {np.min(ratings):.2f}")
print(f"Máximo: {np.max(ratings):.2f}")
print(f"Amplitud (range): {np.ptp(ratings):.2f}")

print("\n=== Votes ===")
print(f"Media: {np.mean(votes):,.0f}")
print(f"Mediana: {np.median(votes):,.0f}")
print(f"Mínimo: {np.min(votes):,.0f}")
print(f"Máximo: {np.max(votes):,.0f}")
print(f"Amplitud (range): {np.ptp(votes):,.0f}")
```

```python
# 4. 
df = df.dropna(subset=['year_of_release', 'movie_rating']).copy()
df['year_of_release'] = df['year_of_release'].astype(int)

# Crear 'década' como entero
df['década'] = ((df['year_of_release'] // 10) * 10).astype(int)

# Rango completo 1920..2020
start_decade = 1920
end_decade   = (df['year_of_release'].max() // 10) * 10
all_decades  = np.arange(start_decade, end_decade + 10, 10, dtype=int)

# Películas por década
pelis_decada = (
    df.groupby('década').size()
      .reindex(all_decades, fill_value=0)
      .rename_axis('década')
      .reset_index(name='cantidad_peliculas')
      .astype({'década':'int','cantidad_peliculas':'int'})
)

# Promedio por década
rating_decada = (
    df.groupby('década')['movie_rating']
      .apply(lambda s: np.mean(s.to_numpy()))
      .reindex(all_decades)
      .rename('promedio_rating')
      .reset_index()
)
rating_decada['promedio_rating'] = rating_decada['promedio_rating'].round(2)

# Resumen
resumen_decada = pelis_decada.merge(rating_decada, on='década', how='left')
resumen_decada = resumen_decada.astype({'década':'int','cantidad_peliculas':'int'})
display(resumen_decada)

# ==============================
# Top 10 películas mejor valoradas
# ==============================
top10 = (
    df.sort_values(by=['movie_rating','votes'], ascending=[False, False])
      .loc[:, ['movie_name', 'year_of_release', 'movie_rating']]
      .head(10)
      .rename(columns={'movie_name':'título','year_of_release':'año','movie_rating':'rating'})
)

# Asegurar tipos visualmente limpios
top10['año'] = top10['año'].astype(int)
top10['rating'] = top10['rating'].astype(float)

display(top10)
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

bios_path = Path("bios.csv")
bios_out  = Path("bios_clean.csv")

# --- Carga (forzar utf-8 para reconocer '•') ---
bios = pd.read_csv(bios_path, encoding="utf-8")
bios.columns = bios.columns.str.strip()

# ===== 1) Eliminar viñetas '•' en Used name (y Roles por si acaso) =====
# Detecta variantes de nombre de columna
def find_col(df, targets):
    targets = {t.lower(): t for t in targets}
    for c in df.columns:
        if c.strip().lower() in targets:
            return c
    return None

used_col  = find_col(bios, {"Used name", "Used names"})
roles_col = find_col(bios, {"Roles"})

if used_col:
    bios[used_col] = (bios[used_col]
                      .astype("string")
                      .str.replace("•", " ", regex=False)
                      .str.replace(r"\s+", " ", regex=True)
                      .str.strip())

if roles_col:
    bios[roles_col] = (bios[roles_col]
                       .astype("string")
                       .str.replace("•", " ", regex=False)
                       .str.replace(r"\s+", " ", regex=True)
                       .str.strip())

# ===== 2) Separar altura y peso desde Measurements =====
meas_col = find_col(bios, {"Measurements"})
if meas_col:
    hw = bios[meas_col].astype("string").str.replace(",", ".", regex=False)
    # antes del "/" -> altura ; después del "/" -> peso
    h_raw = hw.str.split("/", n=1, expand=True)[0].fillna("")
    w_raw = hw.str.split("/", n=1, expand=True)[1].fillna("")

    # extrae primer número (admite decimales)
    height_cm = pd.to_numeric(h_raw.str.extract(r"(\d+(?:\.\d+)?)", expand=False), errors="coerce")
    weight_kg = pd.to_numeric(w_raw.str.extract(r"(\d+(?:\.\d+)?)", expand=False), errors="coerce")

    # si la altura parece estar en metros (<3), conviértela a cm
    height_cm = height_cm.where(~(height_cm < 3), height_cm * 100)

    bios["height_cm"] = height_cm
    bios["weight_kg"] = weight_kg

# ===== 3) Fechas Born/Died -> born_date/died_date =====
date_pat = (
    r"^("
    r"\d{4}-\d{2}-\d{2}|"
    r"[A-Za-z]{3,9}\s+\d{1,2},\s*\d{4}|"
    r"\d{1,2}\s+[A-Za-z]{3,9}\s+\d{4}|"
    r"\d{4}"
    r")"
)

if "Born" in bios.columns:
    born_str = bios["Born"].astype("string").str.strip()
    bios["born_date"] = pd.to_datetime(born_str.str.extract(date_pat, expand=False), errors="coerce")

if "Died" in bios.columns:
    died_str = bios["Died"].astype("string").str.strip()
    bios["died_date"] = pd.to_datetime(died_str.str.extract(date_pat, expand=False), errors="coerce")

# ===== 4) Lugar de nacimiento -> city/region/country =====
if "Born" in bios.columns:
    born = bios["Born"].astype("string").str.strip()

    place_after_in = born.str.extract(r"\bin\s+(.+)$", expand=False)
    born_no_date   = born.str.replace(date_pat, "", regex=True).str.lstrip(", ").str.strip()
    place = place_after_in.fillna(born_no_date).fillna("")

    country_paren  = place.str.extract(r"\(([^)]+)\)\s*$", expand=False)
    place_no_paren = place.str.replace(r"\([^)]+\)\s*$", "", regex=True).str.strip()

    parts = place_no_paren.str.split(",", n=2, expand=True)  # city, region, rest
    bios["born_city"]    = parts[0].str.strip()
    bios["born_region"]  = parts[1].str.strip()
    bios["born_country"] = country_paren.fillna(parts[2].str.strip())
    bios[["born_city","born_region","born_country"]] = bios[["born_city","born_region","born_country"]].replace(["", "nan"], np.nan)

# ===== 5) Eliminar columnas adicionales =====
drop_cols = [c for c in ["Measurements", "Height/Weight", "Height", "Weight"] if c in bios.columns]
bios = bios.drop(columns=drop_cols)

# ===== 6) Guardar =====
bios.to_csv(bios_out, index=False, encoding="utf-8")
print(f"Guardado: {bios_out.resolve()}")

# ===== 7) Mostrar 3 filas original vs limpio =====
print("\n--- Primeros 3 registros del archivo original ---")
display(pd.read_csv(bios_path, encoding="utf-8").head(3))

print("\n--- Primeros 3 registros del archivo limpio ---")
display(pd.read_csv(bios_out, encoding="utf-8").head(3))
```

```python
import pandas as pd
import numpy as np
from pathlib import Path
import warnings
warnings.simplefilter("ignore", pd.errors.DtypeWarning)

# Rutas
res_path = Path("results.csv")
res_out  = Path("results_clean.csv")

# --- Leer original (para la 1ra tabla) ---
res_orig = pd.read_csv(res_path, encoding="utf-8", low_memory=False)
res_orig.columns = res_orig.columns.str.strip()

# --- Copia para limpiar (2da tabla/archivo) ---
res = res_orig.copy()
res.columns = res.columns.str.strip()

#  Games -> Year (año) + Game_type
if "Games" in res.columns:
    g = res["Games"].astype("string").str.strip()
    res["year"]      = pd.to_numeric(g.str.extract(r"(\d{4})", expand=False), errors="coerce").astype("Int64")
    res["game_type"] = g.str.replace(r"^\s*\d{4}\s*", "", regex=True).str.strip()

#  Empates ('=' al inicio) y posición numérica
pos_col = "Position" if "Position" in res.columns else ("Pos" if "Pos" in res.columns else None)
if pos_col:
    ps = res[pos_col].astype("string").str.strip()

    # máscara booleana (NaN -> False) y columna Tied textual
    mask = ps.str.startswith("=").fillna(False)
    res["Tied"] = np.where(mask, "Empate", "No empate")

    # extraer número (p.ej. '=17' -> 17; 'DNS' -> NaN) y dejar como entero nullable
    res["pos"] = pd.to_numeric(ps.str.extract(r"([-+]?\d+)", expand=False), errors="coerce").astype("Int64")

#  Eliminar columnas Unnamed:*
res = res[[c for c in res.columns if not c.lower().startswith("unnamed")]]

for drop_c in ["Games", "Pos", "Position"]:
    if drop_c in res.columns:
        res = res.drop(columns=[drop_c])

#  Reordenar columnas 
preferred_order = [
    "Athlete","NOC","Sport","Discipline","Event",
    "year","game_type","City","Medal","Tied","pos",
    "Team","As","athlete_id","Nationality"
]
order = [c for c in preferred_order if c in res.columns] + [c for c in res.columns if c not in preferred_order]
res = res[order]

#  Nombres iniciando en mayúscula
res.columns = [c[:1].upper() + c[1:] if c else c for c in res.columns]

res.to_csv(res_out, index=False, encoding="utf-8")
display(res_orig.head(10))
display(res.head(10))
```
