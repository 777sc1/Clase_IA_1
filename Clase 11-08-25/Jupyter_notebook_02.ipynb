```python
#Indexacion con columnas
#DATAFRAME> 7 valores de filas y 4 atributos [A,B,C,D]
df2=pd.DataFrame({
        'a':range(7),
        'b':range(7,0,-1),
        'c':['one','one','two','one','two','two','one'],
        'd':[0,1,2,0,1,3,2]
})
#set_index:
df2.set_index(["c","d"])
```

Combinar dataframe

```python
#merge, concat, combine
df1=pd.DataFrame({"key":["b","b","a","c","a","a","b"],
                  "data1":pd.Series(range(7),dtype='Int64')
                  })

df2=pd.DataFrame({"key":["a","b","d"],
                  "data2":pd.Series(range(3),dtype='Int64')
                  })
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
#data1:c, data 2:d
pd.merge(df1,df2,how="outer")
```

```python
izq=pd.DataFrame({'key1':["pri","pri","seg"],
                  'key2':["uno","dos","uno"],
                  'iz_val':pd.Series([1,2,3],dtype='Int64'),
                  })

der=pd.DataFrame({'key1':["pri","pri","seg","seg"],
                  'key2':["uno","uno","uno","dos"],
                  'r_val':pd.Series([4,5,6,7],dtype='Int64'),
                  })

pd.merge(izq,der,on=["key1","key2"],how="outer")
```

```python
#Fusionar tablas por medio de indice
izq1=pd.DataFrame({'key1':["a","b","a","a","b","c"],
                  "valor":pd.Series(range(6),dtype='Int64')
                  })

der1=pd.DataFrame({'g_val':[3.5,7]},
                  index=["a","b"]
                  )
pd.merge(izq1,der1,left_on="key1",right_index=True,how="outer")
```

```python
izqh=pd.DataFrame({"key1":["Curso1","Curso1","Curso1",
                             "Curso2","Curso2"   ],
                   "key2":[2021,2022,2023,2021,2022],
                   "data":pd.Series(range(5),dtype='Int64')
                   })

derh_index=pd.MultiIndex.from_arrays([["Curso2","Curso2","Curso1","Curso1"
                                ,"Curso1","Curso1"],
                               [2022,2021,2021,2021,2022,2023
                                ]])
derh=pd.DataFrame({"notas_1":pd.Series([12,15,16,17,14,18],dtype='Int64',
                                       index=derh_index),
                    "notas_2":pd.Series([10,20,13,15,14,16],dtype='Int64',
                                       index=derh_index),                      
                                       })
pd.merge(izqh,derh,left_on=["key1","key2"],right_index=True)
```

```python
dfo=pd.DataFrame([[7,8],[9,10],[11,12],[16,17]],
                 index=["a","c","e","f"],columns=["Ciudad_1","Ciudad_2"])
dfr=pd.DataFrame([[7,8],[9,10],[11,12],[13,14]],
                 index=["b","c","d","e"],columns=["Ciudad_3","Ciudad_4"])

dfo.join([dfr])
```
